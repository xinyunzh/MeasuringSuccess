```python
import pandas as pd
import numpy as np
```

```python
bundleservice = pd.read_csv("bundledservices.csv",sep =",",header =0)
```

```python
bundleservice.shape
```

```python
bundleservice=bundleservice.sort_values(by = ['clientid','baseline_baseline','quarter'],ascending =[True,False,True])
```

```python
list(bundleservice)
```

```python
bundleservice['AI']=np.where(bundleservice['accessincome']=='Yes',1,0)
```

```python
bundleservice['AE']=np.where(bundleservice['accessemployment']=='Yes',1,0)
```

```python
bundleservice['AF']=np.where((bundleservice['accessfinancialcoach']=='Yes')|(bundleservice['accessfinancialcoach']=="1"),1,0)
```

```python
bundleservice['AF'].value_counts()
```

```python
bundleservice['clientid'] = bundleservice['clientid'].astype(str)
```

```python
bundleservice['accessincome_alltime'] = 0
bundleservice['accessemployment_alltime'] = 0
bundleservice['accessfinancialcoach_alltime'] = 0
```

```python
temid = '0'
AI = 0
AE = 0
AF = 0
for index, row in bundleservice.iterrows():
    if row['clientid'] != temid:
        temid = row['clientid']
        bundleservice.loc[index, 'accessincome_alltime'] = row['AI']
        bundleservice.loc[index, 'accessemployment_alltime'] = row['AE']
        bundleservice.loc[index, 'accessfinancialcoach_alltime'] =row['AF']
        AI = bundleservice.loc[index, 'accessincome_alltime']
        AE = bundleservice.loc[index, 'accessemployment_alltime']
        AF = bundleservice.loc[index, 'accessfinancialcoach_alltime']
    else:
        if AI == 1: 
            bundleservice.loc[index, 'accessincome_alltime'] = 1
        else:
            bundleservice.loc[index,'accessincome_alltime'] =AI + row['AI']
            AI = bundleservice.loc[index, 'accessincome_alltime']
        if AE == 1:
            bundleservice.loc[index, 'accessemployment_alltime'] = 1
        else:
            bundleservice.loc[index, 'accessemployment_alltime'] = AE + row['AE']
            AE = bundleservice.loc[index, 'accessemployment_alltime']
        if AF == 1:
            bundleservice.loc[index, 'accessfinancialcoach_alltime'] = 1
        else:
            bundleservice.loc[index, 'accessfinancialcoach_alltime'] =AF + row['AF']
            AF = bundleservice.loc[index, 'accessfinancialcoach_alltime']
            
            
        
```

```python
bundleservice[['clientid','quarter','baseline_baseline','accessincome', 'accessemployment', 'accessfinancialcoach','AI', 'AE', 'AF', 'accessincome_alltime',
       'accessemployment_alltime', 'accessfinancialcoach_alltime']]
```

```python
bundleservice = bundleservice.drop(['AI','AE','AF'],axis = 1)
```

```python
bundleservice["alltime"] = bundleservice['accessincome_alltime']+bundleservice['accessemployment_alltime']+bundleservice['accessfinancialcoach_alltime']
```

```python
list(bundleservice)
```

```python
bundletable = bundleservice[[ 'clientid',
 'quarter','baseline_baseline','accessincome_alltime',
 'accessemployment_alltime',
 'accessfinancialcoach_alltime',
 'alltime']]
```

```python
bundletable = bundletable.groupby(['clientid']).tail(1)[['clientid','quarter','accessincome_alltime',
 'accessemployment_alltime',
 'accessfinancialcoach_alltime','alltime']]
```

```python
bundletable = bundletable.rename(columns={'accessincome_alltime':'AccessIncome',
                           'accessemployment_alltime':'AccessEmployment',
                           'accessfinancialcoach_alltime':'AccessFinancialCoach',
                           'alltime':'BundledServices'})
#no assign if inplace = True
```
